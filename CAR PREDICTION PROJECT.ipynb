```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import ipywidgets as widgets
from IPython.display import display
```

```python
# Here I am importing my data 
car_data = pd.read_csv('USA_cars_datasets 2.csv')
```

```python
#Viewing the data 
car_data.head()
```

```python
car_data.info()
```

```python
# Here I am dropping some columns I will not need 
car_data.drop(["vin", "lot", "state", "country", "condition"], axis = 1, inplace = True)
```

```python
# Shows the year data
car_data["year"].plot(kind="hist");
```

```python
car_data.plot(kind='scatter', 
             x='year', 
             y='brand', 
             c='price', 
             figsize=(10, 6));
```

```python
# Here i am turing categorical data into numerical
df = pd.get_dummies(car_data)
df
```

```python
Y = np.array(df['price'])
X = df.drop('price', axis = 1)
X = np.array(X)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    Y, 
                                                    test_size=0.2, random_state=42)

price_model = RandomForestRegressor(n_estimators = 1000, random_state = 42)
price_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
y_preds = price_model.predict(X_test)
print(f"R2 score: {r2_score(Y_test, y_preds)}")
```

```python
def to_dollars(value):
    return "${:,.2f}".format(value)

def handle_click(_button):
    price_predicted = get_price()
    print(f"Price predicted is:", to_dollars(price_predicted))

button = widgets.Button(
    description='Calculate Price!',
    disabled=False,
)
button.on_click(handle_click)

button
```

```python
def get_price() : 
    d = {'year': [year_selected.value], 'mileage': [mileage_selected.value], 'brand': [brand_selected.value], 
    'model': [model_selected.value], 'title_status': [title_status_selected.value], 'color': [color_selected.value]}
    idf = pd.DataFrame(data=d)

    dummies_input = pd.get_dummies(idf)

    df_newdata = dummies_input.reindex(labels=df.columns,axis=1)

    df_newdata.fillna(0, inplace=True)
    df_newdata.drop('price', axis=1, inplace=True)

    df_newdata_array = np.array(df_newdata)
    
    price_predicted = price_model.predict(df_newdata_array)
    return price_predicted[0]
```
